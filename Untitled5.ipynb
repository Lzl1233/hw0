```python
###Q1
import numpy as np
import matplotlib.pyplot as plt

thetas = np.linspace(0, 1, 101)

num_thetas = len(thetas)

prior = np.ones(num_thetas) / num_thetas

observations = [1, 0, 1, 1, 0, 1] 

posterior = prior.copy()
all_posteriors = [posterior] 

for x in observations:
    if x == 1:
        likelihood = thetas
    else:
        likelihood = 1 - thetas
    unnormalized = posterior * likelihood

    posterior = unnormalized / unnormalized.sum()

    all_posteriors.append(posterior)

fig, axes = plt.subplots(len(all_posteriors), 1, figsize=(6, 12), sharex=True)

for i, post in enumerate(all_posteriors):
    axes[i].bar(thetas, post, width=0.01, align='center')
    if i == 0:
        axes[i].set_title(f"Step {i}: Prior (uniform)")
    else:
        axes[i].set_title(f"Step {i}: After observing x = {observations[i-1]}")
    axes[i].set_ylim(0, max(post)*1.2)

plt.xlabel(r"$\theta$")
plt.tight_layout()
plt.show()
```

```python
###Q2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def pr_theta1_less_than_theta2(alpha1, beta1, alpha2, beta2, n_sample=2000):
    samples1 = np.random.beta(alpha1, beta1, size=n_sample)
    samples2 = np.random.beta(alpha2, beta2, size=n_sample)
    
    count = np.sum(samples1 < samples2)
    return count / n_sample


def bayesian_two_armed_bandit(theta1_true=0.4, theta2_true=0.6,
                              alpha1=1, beta1=1,
                              alpha2=1, beta2=1,
                              num_rounds=1000):
    choices = []
    rewards = []
    posterior_params = [(alpha1, beta1, alpha2, beta2)]

    for t in range(num_rounds):
        p_flip_coin2 = pr_theta1_less_than_theta2(alpha1, beta1,
                                                  alpha2, beta2)
        if np.random.rand() < p_flip_coin2:
            chosen_arm = 2
            reward = 1 if np.random.rand() < theta2_true else 0
            alpha2 += reward
            beta2  += (1 - reward)
        else:
            chosen_arm = 1
            reward = 1 if np.random.rand() < theta1_true else 0
            alpha1 += reward
            beta1  += (1 - reward)

        choices.append(chosen_arm)
        rewards.append(reward)
        posterior_params.append((alpha1, beta1, alpha2, beta2))

    return np.array(choices), np.array(rewards), posterior_params

if __name__ == "__main__":
    theta1_true = 0.4
    theta2_true = 0.6
    alpha1, beta1 = 1, 1
    alpha2, beta2 = 1, 1
    num_rounds = 200

    choices, rewards, posterior_params = bayesian_two_armed_bandit(
        theta1_true, theta2_true,
        alpha1, beta1,
        alpha2, beta2,
        num_rounds
    )

    rounds = np.arange(1, num_rounds+1)

    plt.figure(figsize=(10, 6))
    plt.scatter(rounds, choices, c=choices, cmap='viridis', s=10)
    plt.yticks([1,2], labels=["Coin 1", "Coin 2"])
    plt.title("Which coin was chosen at each round")
    plt.xlabel("Round")
    plt.ylabel("Chosen Arm")
    plt.show()

    cumulative_rewards = np.cumsum(rewards)
    avg_rewards = cumulative_rewards / rounds
    plt.figure(figsize=(10, 4))
    plt.plot(rounds, avg_rewards, label="Cumulative Avg Reward", color='red')
    plt.axhline(y=theta1_true, color='gray', linestyle='--', label=f"Theta1_true={theta1_true}")
    plt.axhline(y=theta2_true, color='blue', linestyle='--', label=f"Theta2_true={theta2_true}")
    plt.title("Cumulative Average Reward over time")
    plt.xlabel("Round")
    plt.ylabel("Avg Reward")
    plt.legend()
    plt.show()

    steps_to_plot = [0, 1, 5, 10, 50, 100, 199]
    fig, axes = plt.subplots(len(steps_to_plot), 2, figsize=(8, 3*len(steps_to_plot)))

    for i, step in enumerate(steps_to_plot):
        a1, b1, a2, b2 = posterior_params[step]
        xvals = np.linspace(0,1,200)

        pdf1 = beta(a1, b1).pdf(xvals)
        axes[i,0].plot(xvals, pdf1, 'r-')
        axes[i,0].fill_between(xvals, pdf1, alpha=0.3, color='r')
        axes[i,0].set_title(f"Round={step}, Coin1 Posterior: Beta({a1:.1f}, {b1:.1f})")
        axes[i,0].set_ylim(0, None)

        pdf2 = beta(a2, b2).pdf(xvals)
        axes[i,1].plot(xvals, pdf2, 'b-')
        axes[i,1].fill_between(xvals, pdf2, alpha=0.3, color='b')
        axes[i,1].set_title(f"Round={step}, Coin2 Posterior: Beta({a2:.1f}, {b2:.1f})")
        axes[i,1].set_ylim(0, None)

    plt.tight_layout()
    plt.show()
```

```python
###Q3
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, dirichlet

np.random.seed(42)
N = 50
p_true = [0.2, 0.5, 0.7]

data = []
for _ in range(N):
    x1 = 1 if np.random.rand() < p_true[0] else 0
    x2 = 1 if np.random.rand() < p_true[1] else 0
    x3 = 1 if np.random.rand() < p_true[2] else 0
    data.append((x1, x2, x3))

data = np.array(data)

total_successes = data.sum(axis=0)
print (p_true)
print(total_successes, N,)

alphaA = np.array([1.0, 1.0, 1.0])
betaA  = np.array([1.0, 1.0, 1.0])

for (x1, x2, x3) in data:
    alphaA[0] += x1
    betaA[0]  += (1 - x1)
    alphaA[1] += x2
    betaA[1]  += (1 - x2)
    alphaA[2] += x3
    betaA[2]  += (1 - x3)

print()
print(alphaA)
print(betaA)
alphaB = np.array([1.0, 1.0, 1.0])
for (x1, x2, x3) in data:
    alphaB += [x1, x2, x3]

print()
print(alphaB)

fig, axes = plt.subplots(1, 3, figsize=(12, 3))
xgrid = np.linspace(0, 1, 200)
for i in range(3):
    pdf_vals = beta(alphaA[i], betaA[i]).pdf(xgrid)
    axes[i].plot(xgrid, pdf_vals, label=f"Beta({alphaA[i]:.1f},{betaA[i]:.1f})")
    axes[i].fill_between(xgrid, pdf_vals, alpha=0.3)
    axes[i].set_title(f"Entity {i+1} Posterior (Analysis A)")
    axes[i].set_xlabel("p")
    axes[i].legend()
plt.tight_layout()
plt.show()

num_samps = 3000
samps = np.random.dirichlet(alphaB, size=num_samps)

x_coords = samps[:,0] + 0.5 * samps[:,1]
y_coords = (np.sqrt(3)/2) * samps[:,1]

plt.figure(figsize=(5, 5))
plt.scatter(x_coords, y_coords, s=5, alpha=0.3, c='blue')
plt.title(f"Entity1={alphaB[0]:.1f}, Entity2={alphaB[1]:.1f}, Entity3={alphaB[2]:.1f}\nDirichlet Posterior (Analysis B)")
plt.axis('equal')
plt.xticks([])
plt.yticks([])
plt.show()

"""
Conclusion (in English):

- If each trial's successes are distributed among the arms according to
  a multinomial sampling mechanism (e.g., if the total number of successes
  per trial is fixed but split across multiple categories), then the
  Dirichlet-Multinomial framework is more suitable.
"""
```
